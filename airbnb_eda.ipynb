### **Data Analysis and Visualization**

**Ques 1 : Top 5 hosts and their total listings in Airbnb**

```python
# analyzing hosts listings with neighbourhood
host_areas = airbnb_new.groupby(['host_name','neighbourhood_group','host_id'])['calculated_host_listings_count'].max().reset_index()
top_hosts = host_areas.sort_values(by=['calculated_host_listings_count'],ascending = False).head(5)
top_hosts
```

from above we conclude that host name **Sonder(NYC) with host id 219517861	 has most listings i.e 327** in **Manhattan**

```python
#visualizing the hosts with most listing

# Plotting the bar graph
plt.figure(figsize=(8, 5))
plt.bar(top_hosts['host_name'], top_hosts['calculated_host_listings_count'])
plt.xlabel('Host Name')
plt.ylabel('Calculated Host Listings Count')
plt.title('Top Hosts by Calculated Host Listings Count')
plt.xticks(rotation=45)
plt.show()
```

**Ques 2 : How many airbnb are listed in neighbourhood group**

```python
# Number of Airbnb listed in each neighbourhood group
data=airbnb_new.groupby([ 'neighbourhood_group' ]) ['id'].count().reset_index().rename(columns={'id':'count'}).sort_values(by='count', ascending =False)
data.head()
```

```python
# Plotting the bar chart for better visualization
plt.figure(figsize=(8, 5))
plt.bar(data['neighbourhood_group'], data['count'])
plt.xlabel('Neighbourhood Group')
plt.ylabel('Number of Listings')
plt.title('Number of Listings per Neighbourhood Group')
plt.xticks(rotation=45)
plt.show()
```

```python
#checking outliers in price column through box plot
sns.boxplot( x= airbnb_new['price'])
plt.show()
```

```python
# writing a outlier functon to remove outlier from price column
Q1 = airbnb_new['price'].quantile(0.25)
Q3 = airbnb_new['price'].quantile(0.75)
IQR = Q3 - Q1
upper_limit = Q3 +(1.5 * IQR)
lower_limit = Q1 -(1.5 * IQR)
```

```python
# Outliers are removed from price column , now checking with boxplot and also shape of new dataframe
airbnb_new[airbnb_new['price']>upper_limit]
airbnb_new[airbnb_new['price']<lower_limit]
new_airbnb =airbnb_new[airbnb_new['price']<upper_limit]
sns.boxplot(x = new_airbnb['price'])
new_airbnb.shape
plt.show()
```

**Ques 4:What is the distribution of room types in different cities**

```python
# Calculate value counts for 'room_type' and 'neighbourhood_group'
room_type_counts = new_airbnb['room_type'].value_counts()
neighborhood_group_counts = new_airbnb['neighbourhood_group'].value_counts()

# Create a pivot table to get separate columns for room types and neighborhood groups
pivot_table_data = pd.pivot_table(new_airbnb,
                                  index='neighbourhood_group',
                                  columns='room_type',
                                  values='id',
                                  aggfunc='count',
                                  fill_value=0)

# Display the value counts and the pivot table
print("Room Type Counts:")
print(room_type_counts)
print("\nNeighborhood Group Counts:")
print(neighborhood_group_counts)
print("\nData in Pivot Table:")
print(pivot_table_data)
```

```python
plt.figure(figsize=(8,6))
plt.title("Room Type on Neighbourhood Group")
sns.countplot(new_airbnb.neighbourhood_group,x=new_airbnb['neighbourhood_group'],hue=new_airbnb.room_type)
plt.show()

```

The above graph shows that the Entire Home/Apartment is listed most near Manhattan, while Private Rooms and Apartments Near Brooklyn are Nearly equal.

**Ques 5: Who are the top neighbourhood by listing counts**
